The 5 most common problems with messy datasets are (1) column headers are values, not variable names, (2) multiple variables are stored in one column, (3) variables are stored in both rows and columns, (4) multiple types of observational units are stored in the same table, and (5) a single observational unit is stored in multiple tables. 2. Categorical variable: For the `./data/sharks.csv` data covered in the lecture, clean the "Type" variable as well as you can, and explain the choices you make.

```python
df = pd.read_csv('./data/sharks.csv', low_memory=False)
df.head()
#df.columns.tolist()
```

```python
df['Type'].value_counts()
```

```python
type = df['Type'] # create a temporary vector of values for the Type variable to make modifications to

type = type.replace(['Sea Disaster', 'Boat','Boating','Boatomg'],'Watercraft') # consolidate terms related to watercraft/boating into a single category
type.value_counts()

type = type.replace(['Invalid', 'Questionable','Unconfirmed','Unverified','Under investigation'],np.nan) # replaced values that are considered unclean with NaN
type.value_counts()

df['Type'] = type # replace the 'Type' variable with the cleaned version
del type # delete the temporary vector

df['Type'].value_counts()
```

```python
# sharks are much more likely to kill you if the situation is UNprovoked. 
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].replace(['UNKNOWN', 'F','M','2017'],np.nan) # all unclean values
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].replace('y','Y') # all unclean values
pd.crosstab(df['Type'],df['Fatal (Y/N)'],normalize='index')

# The reason why I created a temporary series type that holds the values from the 'Type' column of the dataframe was to 
# allow me to manipulate and clean the data without modifying the original dataframe. Then, I replaced specific lines in 
# the type series to consolidate terms related to watercraft/boating into a single category called 'Watercraft'. This allowed 
# me to standardize categories that represent similar concepts. Next, I replaced values that are considered unclean with NaN. 
# Then, I replaced the original 'Type' column in the dataframe with the cleaned type series to updated it with the standardized 
# and cleaned values. Lastly, I created a table to analyze the distribution of fatal incidents across different types of incidents. 
```

```python
length = df['ImposedSentenceAllChargeInContactEvent']
type = df['SentenceTypeAllChargesAtConvictionInContactEvent']

# print( length.unique()  , '\n') # Some values are ' ', denoting missing
length = pd.to_numeric(length,errors='coerce') # convert to numeric
length_NA = length.isnull() # create a missing dummy
print( np.sum(length_NA),'\n') # 9k missing values of 23k

print( pd.crosstab(length_NA, type), '\n') # Category 4 is cases where the charges were dismissed

length = length.mask( type == 4, 0) # Replace length with 0 when type ==4
length = length.mask( type == 9, np.nan) # Replace length with np.nan when type == 9

length_NA = length.isnull() # Create a new missing dummy
print( pd.crosstab(length_NA, type), '\n')
print( np.sum(length_NA),'\n') # 274 missing

df['ImposedSentenceAllChargeInContactEvent'] = length # replace data with cleaned version
del length, type # delete temporary length/type variables

# Temporary variables of 'length' and 'type' are created to hold the values from the 'ImposedSentenceAllChargeInContactEvent' 
# and 'SentenceTypeAllChargesAtConvictionInContactEvent' columns. This allowed me to manipulate and clean these columns. 'length' 
# is then converted to numeric values, and non-numeric values are coerced into NaN. This organizes the column to contain only numeric 
# or NaN values. 'length'  is updated based on the values in the type column. For rows where type is 4 (charges dismissed), 'length' 
# is replaced with 0. For rows where type is 9, 'length' is set to NaN (indicating missing or irrelevant data). 'length_NA' is then 
# recalculated to reflect the updated length values. 
```
